```python
from qiskit.quantum_info import Statevector, Operator, Pauli
from IPython.display import display
import numpy as np
```

```python
zero, one = Statevector.from_label('0'), Statevector.from_label('1')
psi = zero.tensor(one)
display(psi)

plus, minus = Statevector.from_label('+'), Statevector.from_label('-')
plus_i, minus_i = Statevector.from_label('r'), Statevector.from_label('l')
plus.tensor(minus_i).draw('latex')
```

```python
state1 = Statevector([1+1j, 1/3])
psi = one.tensor(state1).draw()
display(psi.is_valid())
```

```python
from qiskit import QuantumCircuit

psi1 = plus.tensor(plus_i)
qc = QuantumCircuit(2)
qc.x(0)
qc.id(1)
qc.draw()

psi = psi1.evolve(qc)
psi.draw('latex')
```

```python
qc.cx(1,0)
qc.draw()

psi.evolve(qc).draw('latex')
```
